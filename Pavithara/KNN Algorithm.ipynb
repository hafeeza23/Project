```python
target = train_data["Primary Type "].unique()
print(target)
```

```python
test_data = pd.read_csv("/content/drive/MyDrive/Project/Crime (18-p)/Crime - (18-p) 1_final/input/test.csv")
print(test_data.head())
print(test_data.shape)
```

```python
data_dict = {}
count = 1
for data in target:
    data_dict[data] = count
    count+=1
train_data["Primary Type"] = train_data["Primary Type"].replace(data_dict)

#Replacing the day of weeks
data_week_dict = {
    "Monday": 1,
    "Tuesday":2,
    "Wednesday":3,
    "Thursday":4,
    "Friday":5,
    "Saturday":6,
    "Sunday":7
}
train_data["DayOfWeek"] = train_data["DayOfWeek"].replace(data_week_dict)
test_data["DayOfWeek"] = test_data["DayOfWeek"].replace(data_week_dict)
#District
district = train_data["Closest Station"].unique()
data_dict_district = {}
count = 1
for data in district:
    data_dict_district[data] = count
    count+=1 
train_data["Closest Station"] = train_data["Closest Station"].replace(data_dict_district)
test_data["Closest Station"] = test_data["Closest Station"].replace(data_dict_district)
```

```python
features = ["Case Number", "Closest Station",  "Lattitude", "Longtitude"]
X_train = train_data[features]
y_train = train_data["Primary type"]
X_test = test_data[features]

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
```
